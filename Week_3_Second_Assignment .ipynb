```python
import pandas as pd
import numpy as np 
from bs4 import BeautifulSoup
import requests
data = pd.DataFrame(columns=(["Postal Code","Borough","Neighbrough"]))
listA = []
with open('index.html') as f:
    soup = BeautifulSoup(f,"lxml")  
    allDv = soup.find_all('td')
    
    
    
for th in allDv:
    listA.append(th)
   
```

```python
number = 0
for i,k in enumerate(listA):
    if(i%3==0):
        data.set_value(number,"Postal Code",listA[i].text)
    elif(i%3==1):
        data.set_value(number,"Borough",listA[i].text)
    elif(i%3==2):
        data.set_value(number,"Neighbrough",listA[i].text)
        number = number +1 
    
```
